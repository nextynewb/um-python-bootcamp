```python
"""
Basic Plotting: Line Chart
"""

import matplotlib.pyplot as plt

years = [2016, 2017, 2018, 2019, 2020]
inflation_rates = [1.5, 1.8, 1.2, 2.2, 2.7]

plt.plot(years, inflation_rates, marker='o')
plt.xticks(years)
plt.show()
```

```python
"""
Line Chart with more customization
"""
import matplotlib.pyplot as plt

years = [2016, 2017, 2018, 2019, 2020]
inflation_rates = [1.5, 1.8, 1.2, 2.2, 2.7]


# Common parameters for plt.plot:
# - marker: 'o', 's', '^', 'v', '<', '>', 'p', '*', 'h', '+', 'x', 'd', '|', '_'
# - color: 'red', 'blue', 'green', 'yellow', 'black', etc. or hex code like '#FF5733'
# - linestyle: '-' (solid), '--' (dashed), '-.' (dash-dot), ':' (dotted), 'None'
# - linewidth or lw: thickness of the line (float value)
# - alpha: transparency (0.0 to 1.0)
# - label: string for legend

plt.plot(years, inflation_rates, marker='o', color='red', linestyle='--', linewidth=2, label='Inflation Rate')
plt.legend(loc='upper left')
plt.xticks(years)
plt.show()
```

```python
"""
Subplots with multiple rows and columns
"""


import matplotlib.pyplot as plt

years = [2016, 2017, 2018, 2019, 2020]
inflation_rates = [1.5, 1.8, 1.2, 2.2, 2.7]


fig, axs = plt.subplots(2, 2, figsize=(8, 6))

# First row

axs[0,0].plot(years, inflation_rates, color='red')
axs[0,1].bar(years, inflation_rates, color='red')

categories = ['Food', 'Transportation', 'Housing', 'Healthcare', 'Entertainment']
inflation_by_category = [1.5, 1.8, 1.2, 2.2, 2.7]

# Second row

axs[1,0].pie(inflation_by_category, labels=categories, autopct='%1.1f%%', startangle=140)
axs[1,1].bar(categories, inflation_by_category, color='red')

# Fix the rotation issue by setting labels separately
axs[1,1].set_xticklabels(categories, rotation=20, fontsize=8)

plt.show()
```

```python
"""
Adding utils in subplots
"""


import matplotlib.pyplot as plt

years = [2016, 2017, 2018, 2019, 2020]
inflation_rates = [1.5, 1.8, 1.2, 2.2, 2.7]


fig, axs = plt.subplots(2, 2, figsize=(18, 10))

# First row

axs[0,0].plot(years, inflation_rates, color='red')
axs[0,0].set_title('Line Chart')
axs[0,0].set_xlabel('Year')
axs[0,0].set_ylabel('Inflation Rate')

axs[0,1].bar(years, inflation_rates, color='red')
axs[0,1].set_title('Bar Chart')
axs[0,1].set_xlabel('Year')
axs[0,1].set_ylabel('Inflation Rate')

categories = ['Food', 'Transportation', 'Housing', 'Healthcare', 'Entertainment']
inflation_by_category = [1.5, 1.8, 1.2, 2.2, 2.7]

# Second row

axs[1,0].pie(inflation_by_category, labels=categories, autopct='%1.1f%%', startangle=140)
axs[1,0].set_title('Pie Chart')
axs[1,0].set_xlabel('Year')
axs[1,0].set_ylabel('Inflation Rate')

axs[1,1].bar(categories, inflation_by_category, color='red')
axs[1,1].set_title('Bar Chart')
axs[1,1].set_xlabel('Year')
axs[1,1].set_ylabel('Inflation Rate')
axs[1,1].set_xticklabels(categories, rotation=20, fontsize=8)

plt.show()
```

```python
"""
Exercise
"""

df = pd.read_csv('https://raw.githubusercontent.com/nextynewb/um-python-bootcamp/main/dataset/economic_data_2005.csv')
fig, axs = plt.subplots(2, 2, figsize=(15, 15))


df_malaysia = df[(df['country'] == 'Malaysia') & (df['date']>=2010)]
axs[0,0].plot(df_malaysia['date'], df_malaysia['GDP (current US$)'], color='red')
axs[0,0].set_title('GDP of Malaysia')
axs[0,0].set_xticks(df_malaysia['date'])
axs[0,0].set_xticklabels(df_malaysia['date'], rotation=45)


axs[0,1].bar(df_malaysia['date'], df_malaysia['Unemployment, total (% of total labor force)'], color='red')
axs[0,1].set_title('Unemployment of Malaysia')
axs[0,1].set_xticks(df_malaysia['date'])
axs[0,1].set_xticklabels(df_malaysia['date'], rotation=45)


axs[1,0].plot(df_malaysia['date'], df_malaysia['Inflation, consumer prices (annual %)'], color='red')
axs[1,0].set_title('Inflation of Malaysia')
axs[1,0].set_xticks(df_malaysia['date'])
axs[1,0].set_xticklabels(df_malaysia['date'], rotation=45)

axs[1,1].pie(df_malaysia['Unemployment, total (% of total labor force)'], labels=df_malaysia['date'], autopct='%1.1f%%')
axs[1,1].set_title('Unemployment of Malaysia')

plt.show()

```
